## Business Requirements

## KPI Requirements

```python
# Total Sales

Totalsales = df['Sales'].sum()

# Average Sales

Avgsales = df['Sales'].mean()

# No of items sold

itemsold = df['Sales'].count()

# Average Rating

AvgRating = df['Rating'].mean()

# Display Result

print(f"Total Sales      : $ {Totalsales:,.0f}")
print(f"Average Sales    : $ {Avgsales:,.1f}")
print(f"NO Of Items Sold : {itemsold:,.0f}")
print(f"Average Rating   : {AvgRating:,.1f}")
```

## Charts Requirements

#### Total Sales by Fat Content

```python
sales_by_fat = df.groupby('Item Fat Content')['Sales'].sum()

ply.pie(sales_by_fat, labels = sales_by_fat.index,
                      autopct = '%.0f%%',
                      startangle = 90)

ply.title('Sales By Fat Content')
ply.axis('equal')
ply.show()
```

#### Total Sales by Item Type

```python
sales_by_item_type = df.groupby('Item Type')['Sales'].sum().sort_values(ascending=False)

ply.figure(figsize=(10,6))
bars = ply.bar(sales_by_item_type.index, sales_by_item_type.values)

ply.xticks(rotation=-90)
ply.xlabel('Item Type')
ply.ylabel('Total Sales')
ply.title('Total Sales By Items Sold')

for bar in bars:
    ply.text(bar.get_x() + bar.get_width() / 2, bar.get_height(),
             f'{bar.get_height():,.0f}', ha='center', va='bottom', fontsize=8)

ply.tight_layout()
ply.show()
```

#### Fat Content By Outlet For Total Sales

```python
fat_content_by_outlet = df.groupby(['Outlet Location Type','Item Fat Content'])['Sales'].sum().unstack()
fat_content_by_outlet =fat_content_by_outlet[['Regular', 'Low Fat']]

ax = fat_content_by_outlet.plot(kind= 'bar', figsize = (8,5), title = 'Outlet Tier By Item Fat Content')
ply.xlabel('Outlet Location Type')
ply.ylabel('Total Sales')
ply.legend(title='Item Fat Content')

ply.tight_layout()
ply.show()
```

#### Total Sales By Outlet Establishment Year

```python
sales_by_year = df.groupby('Outlet Establishment Year')['Sales'].sum().sort_index()

ply.figure(figsize=(10,5))
ply.plot(sales_by_year.index, sales_by_year.values, marker='o', linestyle='-')

ply.title('Outlet Establishment')
ply.xlabel('Outlet Establishment Year')
ply.ylabel('Total Sales')

for x, y in zip(sales_by_year.index, sales_by_year.values):
    ply.text(x, y, f'{y:,.0f}', ha='center', va='bottom', fontsize=8)


ply.tight_layout()
ply.show()
```

#### Sales By Outlet Size

```python
sales_by_outletsize = df.groupby('Outlet Size')['Sales'].sum()

ply.figure(figsize= (4,4))
ply.pie(sales_by_outletsize, labels = sales_by_outletsize.index,
                             autopct = '%.0f%%',
                             startangle = 90)

ply.title('Outlet Size By Sales')
ply.axis('equal')
ply.tight_layout()
ply.show()
```

#### Sales By Outlet Location

```python
sales_by_outletlocation = df.groupby('Outlet Location Type')['Sales'].sum().reset_index()
sales_by_outletlocation = sales_by_outletlocation.sort_values('Sales', ascending = False)

ply.figure(figsize = (8,4))
axx = sns.barplot(x = 'Sales', y = 'Outlet Location Type', data = sales_by_outletlocation)

ply.title('Total Sales By Outlet Location Type')
ply.ylabel('Outlet Location Type')
ply.xlabel(' Total Sales')

ply.tight_layout()
ply.show()
```
